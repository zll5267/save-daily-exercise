```python


num_samples = 320 # 20ms

frequency_s = 3000 # number of times a wave repeats a second
sine_wave_s = [np.sin(2 * np.pi * frequency_s * x/sampling_rate) for x in range(num_samples)]

frequency_0 = 5000 # number of times a wave repeats a second
sine_wave_0 = [np.sin(2 * np.pi * frequency_0 * x/sampling_rate) for x in range(num_samples)]

frequency_1 = 5100 # number of times a wave repeats a second
sine_wave_1 = [np.sin(2 * np.pi * frequency_1 * x/sampling_rate) for x in range(num_samples)]

frequency_2 = 5200 # number of times a wave repeats a second
sine_wave_2 = [np.sin(2 * np.pi * frequency_2 * x/sampling_rate) for x in range(num_samples)]

frequency_3 = 5300 # number of times a wave repeats a second
sine_wave_3 = [np.sin(2 * np.pi * frequency_3 * x/sampling_rate) for x in range(num_samples)]

frequency_4 = 5400 # number of times a wave repeats a second
sine_wave_4 = [np.sin(2 * np.pi * frequency_4 * x/sampling_rate) for x in range(num_samples)]

frequency_5 = 5500 # number of times a wave repeats a second
sine_wave_5 = [np.sin(2 * np.pi * frequency_5 * x/sampling_rate) for x in range(num_samples)]
```

```python
nframes=num_samples
comptype="NONE"
compname="not compressed"
nchannels=1
sampwidth=2 # short int

wav_file = wave.open(file, 'w')
#print(help(wav_file.setparams))
wav_file.setparams((nchannels, sampwidth, int(sampling_rate), nframes, comptype, compname))

for i in range(10):
    for s in sine_wave_s:
        wav_file.writeframes(struct.pack('h', int(s*amplitude)))

    for s in sine_wave_s:
        wav_file.writeframes(struct.pack('h', int(s*amplitude)))

    for s in sine_wave_0:
        wav_file.writeframes(struct.pack('h', int(s*amplitude)))

    for s in sine_wave_1:
        wav_file.writeframes(struct.pack('h', int(s*amplitude)))

    for s in sine_wave_2:
        wav_file.writeframes(struct.pack('h', int(s*amplitude)))

    for s in sine_wave_3:
        wav_file.writeframes(struct.pack('h', int(s*amplitude)))

    for s in sine_wave_4:
        wav_file.writeframes(struct.pack('h', int(s*amplitude)))

    for s in sine_wave_5:
        wav_file.writeframes(struct.pack('h', int(s*amplitude)))

    for s in sine_wave_s:
        wav_file.writeframes(struct.pack('h', int(s*amplitude)))

    for s in sine_wave_s:
        wav_file.writeframes(struct.pack('h', int(s*amplitude)))
    
wav_file.close()
```
