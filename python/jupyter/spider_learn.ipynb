```python
dayStart = "2018-01-01"
dayEnd = "2018-01-03"
def get_award_info_between_date(start, end):
    return "http://www.cwl.gov.cn/cwl_admin/kjxx/findDrawNotice?name=ssq&issueCount=&issueStart=&issueEnd=&dayStart=" + start + "&dayEnd="+ end + "&pageNo="
```

```python
lottery_request = get_award_info_between_date(dayStart, dayEnd)
lottery_request
```

```python
lottery_request_headers = {"X-Requested-With":"XMLHttpRequest", 
                           "Referer":"http://www.cwl.gov.cn/kjxx/ssq/kjgg/",
                           "User-Agent":"Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/57.0.2987.110 Safari/537.36",
                          }
```

```python
lottery_res = requests.get(lottery_request, headers=lottery_request_headers)
lottery_res, lottery_res.text
```

```python
lottery_json = json.loads(lottery_res.text)
lottery_json
```

```python
lottery_results = []
for item in lottery_json['result']:
    entry = {}
    print(item["code"],item["red"],item["blue"])
    entry["code"] = item["code"]
    entry["red"] = item["red"]
    entry["blue"] = item["blue"]
    lottery_results.append(entry)
```

```python
import pandas
lottery_df = pandas.DataFrame(lottery_results)
lottery_df.head()
```

```python
lottery_df.to_csv("lottery.csv")
```

```python
import sqlite3
with sqlite3.connect("lottery.sqlite") as db:
    lottery_df.to_sql("lottery", con=db)
```
